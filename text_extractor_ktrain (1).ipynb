```python
def download_pdf_new(url):
  ''' downloads pdf from url in series and prints if pdf not available'''
  try:
    wget.download(url, out = '/content/models_ex')
  except:
    print("An exception occurred")

```

```python
# file path for the corpus documents
docs = '/content/models_ex'
```

```python

ktrain.text.SimpleQA.index_from_folder(docs, INDEXDIR, commit_every=len(docs),
                              multisegment=True, procs=4, # these args speed up indexing
                               breakup_docs=True,         # this slows indexing but speeds up answer retrieval
                              use_text_extraction=True)
```

```python
qa = ktrain.text.SimpleQA(INDEXDIR)
```

```python
answers = qa.ask('price of baobab')
qa.display_answers(answers)
```

```python
answers = qa.ask('value of baobab powder')
qa.display_answers(answers)
```
